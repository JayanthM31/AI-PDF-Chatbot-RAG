```python
# Install required libraries
!pip install sentence-transformers faiss-cpu

# Import libraries
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np

# Step 1: Realistic dataset
documents = [
    "Python is widely used for machine learning and artificial intelligence",
    "Deep learning uses neural networks for complex pattern recognition",
    "FAISS is a library for efficient similarity search and clustering",
    "Data science involves statistics, programming, and data analysis",
    "Automobile engines require regular maintenance and repair",
    "Machine learning models can detect fraud in financial transactions",
    "Natural language processing helps computers understand human language",
    "Big data technologies include Hadoop and Spark",
    "AI systems can be used for recommendation engines",
    "Electrical circuits need proper wiring and safety precautions"
]

# Step 2: Load embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Step 3: Convert documents to embeddings
doc_embeddings = model.encode(documents)

# Step 4: Create FAISS index
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(doc_embeddings))

# Step 5: Semantic search function (FINAL VERSION)
def semantic_search(query, k=3):
    query_embedding = model.encode([query])
    D, I = index.search(np.array(query_embedding), k)

    print("\n==============================")
    print("Query:", query)
    print("==============================")

    print("\nUnderstanding your query using semantic embeddings...\n")
    print("Top Results:\n")

    for i in range(k):
        # Convert distance to relevance score
        score = 1 / (1 + D[0][i])

        # Assign match level
        if score > 0.5:
            level = "High Match"
        elif score > 0.3:
            level = "Medium Match"
        else:
            level = "Low Match"

        print(f"{i+1}. {documents[I[0][i]]}")
        print(f"   Relevance Score: {score:.4f}")
        print(f"   Match Level: {level}\n")

    print("Search completed successfully!")

# Step 6: Interactive loop
while True:
    query = input("\nEnter your search query (type 'exit' to stop): ")

    if query.lower() == "exit":
        print("Search stopped.")
        break

    semantic_search(query)
```
